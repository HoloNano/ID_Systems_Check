One figure that I am learning how to create and demonstrate is a stem plot. I think this is very similar to a bar graph; however, due to the sheer amount of data in this csv file, a stem plot would (hopefully) allow me to display the data without the graph looking visually claustrophobic.

```python
import numpy as np
```

```python
fig1, ax1 = plt.subplots()
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))
plt.stem(x, y)
plt.show()
```
