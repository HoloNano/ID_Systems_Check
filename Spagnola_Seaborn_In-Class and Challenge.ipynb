```python
df_salary = pd.read_csv("Salary Data.csv")
df_salary
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", data=df_salary)
plt.title('Default Scatter Plot: Age vs Salary')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=df_salary)
plt.title('Default Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=df_salary, palette=education_colors)
plt.title('Default Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=df_salary, palette=education_colors, markers=gender_markers)
plt.title('Default Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}
plt.figure()
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=df_salary, palette=education_colors, markers=gender_markers)
plt.title('Default Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}
plt.figure(figsize=(12, 8))
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=df_salary, palette=education_colors, markers=gender_markers)
plt.legend(ncol=2, frameon=False, loc= "lower right" )
plt.title('Default Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python

gender_colors = {"Female": "red", "Male": "blue"}
plt.figure(figsize=(10, 8))
sns.set_style("ticks")
Bachelors=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=df_salary[df_salary["Education Level"] == "Bachelor's"], palette=gender_colors)
Bachelors.set_xlim(25, 50)
Bachelors.set_ylim(50000, 250000)
plt.legend(frameon=False, loc= "upper left" )
plt.title("Default Scatter Plot: Age vs Salary Based on Gender Among Bachelor's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}
plt.figure(figsize=(10, 8))
sns.set_style("ticks")
Masters=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=df_salary[df_salary["Education Level"] == "Master's"], palette=gender_colors)
Masters.set_xlim(25, 50)
Masters.set_ylim(50000, 250000)
plt.legend(frameon=False, loc= "upper left" )
plt.title("Default Scatter Plot: Age vs Salary Based on Gender Among Masters's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}
plt.figure(figsize=(10, 8))
sns.set_style("ticks")
PhD=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=df_salary[df_salary["Education Level"] == "PhD"], palette=gender_colors)
PhD.set_xlim(25, 50)
PhD.set_ylim(50000, 250000)
plt.legend(frameon=False, loc= "upper left" )
plt.title("Default Scatter Plot: Age vs Salary Based on Gender Among PhD Degree Holders")
plt.show()
```
