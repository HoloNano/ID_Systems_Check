```python
# Importing Pandas, Matplotlib, and numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing the newly installed plotly library
import plotly.express as px
```

```python
# Reading our Pokemon csv data and assigning '?' values with NaN for tidying
pokemon = pd.read_csv('pokedex.csv', na_values = '?')
pokemon.head()
```

```python
#Creating a new variable for our tidied data. For looking at catch rates and stat totals, we do not need these columns at the moment.

pokemon_tidy = pokemon.drop(columns = ['Type II', \
                                       'Ability II', \
                                       'Hidden Ability', \
                                       'EV Worth', \
                                       'Gender', \
                                       'Egg Group I', \
                                       'Egg Group II', \
                                       'Evolve']).dropna()
pokemon_tidy
                                       
```

```python
# Creating our first scatter plot with this data. Observing Pokemon catch rates with their stat totals based on their typings

fig = px.scatter(pokemon_tidy, x = 'Catch', y = 'Total',
                 color = 'Type I',
                 title = 'Pokemon Stats Organized by Type')
fig.show()
```

```python
#In order to have all of our values work with a bubble chart, 
#we need to fill in the NaN values with a placeholder so most of our data can be recognized and read

pokemon_tidy.fillna(0)
pokemon_tidy
```

```python
# Creating a bubble chart based on the initial scatter plot observing their names, stat totals, typing, and catch rates
# This chart can be used by players to determine their odds of catching powerful pokemon for their teams

fig = px.scatter(pokemon_tidy, x = 'Catch', y = 'Total',
                 color = 'Type I',
                 size = 'Catch',
                 title = 'Pokemon Organized by Catch Rate and Stat Totals',
                 width = 1000, height = 750,
                 hover_name = 'Pokemon',
                color_discrete_map = {'Grass': 'lightgreen',
                                      'Ghost': 'lightgrey',
                                      'Dark': 'navy',
                                      'Dragon': 'mediumpurple',
                                      'Fighting': 'firebrick',
                                      'Electric': 'gold',
                                      'Water': 'aqua',
                                      'Poison': 'darkviolet',
                                      'Bug': 'forestgreen',
                                      'Ground': 'peru',
                                      'Normal': 'dodgerblue',
                                      'Psychic': 'lightsteelblue',
                                      'Ice': 'paleturquoise',
                                      'Fire': 'orangered',
                                      'Rock': 'dimgrey',
                                      'Fairy': 'orchid',
                                      'Steel': 'silver',
                                      'Flying': 'lightcyan'})
fig.show()
```
